```python
from collections import defaultdict
# reverse the antibody chain
for dir in os.listdir('/home/yubeen/alphafold2.3_monomer'):
    if (len(dir.split('_')) < 4) or ('.' in dir):
        continue
    # if dir.split('_')[2] == '#': continue
    ab_block_path = f"/home/yubeen/alphafold2.3_monomer/{dir}/{dir}_ab/ranked_0.pdb"
    print(ab_block_path)
    with open(ab_block_path, 'r') as file:
        lines = file.readlines()

    chains = defaultdict(list)
    for line in lines:
        if line.startswith("ATOM"):
            chain = line[21]
            chains[chain].append(line)
    print("Chains", chains.keys())
    reversed_lines = []
    for chain in sorted(chains.keys(), reverse=False):
        print("Reversing chain", chain)
        reversed_lines.extend(chains[chain])

    new_ab_block_path = f"/home/kkh517/alphafold2.3_ab_benchmark/{dir}/new_ab_block.pdb"
    with open(new_ab_block_path, 'w') as file:
        file.writelines(reversed_lines)

    print(f"Reversed antibody chain saved to {new_ab_block_path}")
```
